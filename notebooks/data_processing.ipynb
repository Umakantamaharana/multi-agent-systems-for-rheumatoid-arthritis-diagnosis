```python
import pandas as pd
import glob
import os

# Define the directory containing the CSV files
directory = "results/two_agent_kb"

# Get all CSV files in the directory
csv_files = glob.glob(os.path.join(directory, "*.csv"))

# Function to clean the <think>...</think> part from a string
def clean_think_tags(value):
    if isinstance(value, str):
        # Find the position of </think> and keep the part after it
        end_tag = "</think>"
        if end_tag in value:
            return value.split(end_tag, 1)[-1].strip()
    return value

# Process each CSV file
for file in csv_files:
    # Read the CSV file
    print(file)
    df = pd.read_csv(file)
    
    # Apply the cleaning function to all values in the DataFrame
    df = df.applymap(clean_think_tags)
    
    # Save the cleaned DataFrame back to the same file or a new file
    cleaned_file = file.replace(".csv", "_cleaned.csv")
    df.to_csv(cleaned_file, index=False)
    print(f"Cleaned file saved: {cleaned_file}")
```
